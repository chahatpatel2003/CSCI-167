```python
# -----------------------
# 1. Import required libraries
# -----------------------
import numpy as np
import matplotlib.pyplot as plt

# -----------------------
# 2. Create input values from -2 to 2
# -----------------------
x = np.linspace(-2, 2, 100)

# -----------------------
# 3. Define basis functions (phi_0 to phi_3)
# Each returns a vector (same shape as x)
# -----------------------

def phi_0(x):
    return np.ones_like(x)   # Constant 1 for all x

def phi_1(x):
    return x                 # Linear

def phi_2(x):
    return np.maximum(0, x - 0.5)   # ReLU shifted right

def phi_3(x):
    return np.maximum(0, -x - 0.5)  # ReLU mirrored & shifted left

# -----------------------
# 4. Plot each phi function
# -----------------------
plt.figure(figsize=(10, 6))
plt.plot(x, phi_0(x), label='phi_0(x)')
plt.plot(x, phi_1(x), label='phi_1(x)')
plt.plot(x, phi_2(x), label='phi_2(x)')
plt.plot(x, phi_3(x), label='phi_3(x)')
plt.title("Basis Functions")
plt.xlabel("x")
plt.ylabel("phi(x)")
plt.legend()
plt.grid(True)
plt.show()

# -----------------------
# 5. Choose weights (psi values)
# -----------------------
psi_0 = 0.5
psi_1 = 1.0
psi_2 = -1.0
psi_3 = 0.5

# -----------------------
# 6. Combine the basis functions with weights
# -----------------------
y = (
    psi_0 * phi_0(x) +
    psi_1 * phi_1(x) +
    psi_2 * phi_2(x) +
    psi_3 * phi_3(x)
)

# -----------------------
# 7. Plot the final result y(x)
# -----------------------
plt.figure(figsize=(8, 5))
plt.plot(x, y, color='green', label='y(x)')
plt.title("Final Output Function: y(x)")
plt.xlabel("x")
plt.ylabel("y(x)")
plt.grid(True)
plt.legend()
plt.show()

# -----------------------
# 8. Answer the To-Do Questions
# -----------------------
print("\nTo-Do Answers:")

print("1. Increasing psi_1 (the weight on x) makes the sloped region steeper.")
print("2. Multiplying psi_2, psi_2.1, psi_2.2, psi_2.3 (clipped functions) by -1 flips those parts vertically.")
print("3. Setting psi_3 to -1 will invert the phi_3(x) shape (the part that activates for x < -0.5).")
```
