```python

import numpy as np
import matplotlib.pyplot as plt
import math


def number_regions(Di, D):
    N = 0
    for j in range(0, min(Di, D) + 1):
        N += math.comb(D, j)
    return N


N = number_regions(2, 3)
print(f"Di=2, D=3, Number of regions = {int(N)}, True value = 7")


N = number_regions(10, 50)
print(f"Di=10, D=50, Number of regions = {int(N)}, True value = 13432735556")


try:
    N = number_regions(10, 8)
    print(f"Di=10, D=8, Number of regions = {int(N)}, True value = 256")
except Exception as error:
    print("Error:", error)


D = 8; Di = 10
N = np.power(2, D)
print(f"Using 2^D formula: {N}")


dims = np.array([1, 5, 10, 50, 100])
regions = np.zeros((dims.shape[0], 1000))
for c_dim in range(dims.shape[0]):
    D_i = dims[c_dim]
    for D in range(1, 1000):
        regions[c_dim, D] = number_regions(D_i, D)

plt.figure(figsize=(8,6))
for i in range(dims.shape[0]):
    plt.plot(np.arange(1, 1000), regions[i, 1:], label=f"Di={dims[i]}")
plt.yscale("log")
plt.xlabel("Hidden Units D")
plt.ylabel("Number of Regions (log scale)")
plt.legend()
plt.title("Figure 3.9a: Regions vs Hidden Units")
plt.grid(True)
plt.show()


def number_parameters(D_i, D):

    return (D_i + 1) * D + (D + 1)


N = number_parameters(10, 8)
print(f"Di=10, D=8, Number of parameters = {int(N)}, True value = 97")


dims = np.array([1, 5, 10, 50, 100])
regions = np.zeros((dims.shape[0], 200))
params = np.zeros((dims.shape[0], 200))

for c_dim in range(dims.shape[0]):
    D_i = dims[c_dim]
    for D in range(1, 200):
        regions[c_dim, D] = number_regions(D_i, D)
        params[c_dim, D] = number_parameters(D_i, D)

plt.figure(figsize=(8,6))
for i in range(dims.shape[0]):
    plt.plot(params[i, 1:], regions[i, 1:], label=f"Di={dims[i]}")
plt.yscale("log")
plt.xlabel("Number of Parameters")
plt.ylabel("Number of Regions (log scale)")
plt.legend()
plt.title("Figure 3.9b: Regions vs Parameters")
plt.grid(True)
plt.show()
```
