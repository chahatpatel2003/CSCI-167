```python


import numpy as np
import matplotlib.pyplot as plt


def f(x, phi0, phi1):
    return phi0 + phi1 * x


def compute_loss(x, y, phi0, phi1):
    y_pred = f(x, phi0, phi1)
    return np.mean((y - y_pred)**2)


theta0_true, theta1_true = 2.0, 0.5
x = np.linspace(-5, 5, 50)
y = theta0_true + theta1_true * x + np.random.normal(scale=1.0, size=x.shape)

def plot(x, y, phi0, phi1):
    plt.scatter(x, y, label='Data')
    plt.plot(x, f(x, phi0, phi1), 'r-', label='Model')
    plt.xlabel('x'); plt.ylabel('y')
    plt.legend(); plt.show()


phi0, phi1 = 1.6, -0.8
plot(x, y, phi0, phi1)
print(f'Loss = {compute_loss(x,y,phi0,phi1):.3f}')


phi0, phi1 = 0.0, 0.0
alpha = 0.01

for i in range(1000):
    y_pred = f(x, phi0, phi1)
    dphi0 = -2 * np.mean(y - y_pred)
    dphi1 = -2 * np.mean((y - y_pred) * x)
    phi0 -= alpha * dphi0
    phi1 -= alpha * dphi1

print("Optimized phi0, phi1:", phi0, phi1)
print("Final Loss:", compute_loss(x, y, phi0, phi1))
plot(x, y, phi0, phi1)


```
